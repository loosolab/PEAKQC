This is a preprocess-pipeline for singel cell ATACseq data.
As input data fastq or any later datatype(.bam, .narrowPeak, .snap) can be used. The final output is an h5ad file suitable for episcanpy.
This pipeline calls functions from Snaptools and SnapATAC to index the reference genome (snaptools), perform the alignment (snaptools), build the snapfile (snaptools), to build the peak matrix (snaptools) and export a rds file of a snap object (snapatac).
